```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define path for processed data and read
processed_data_path = os.path.join(os.path.pardir, 'data', 'processed')
train_file_path = os.path.join(processed_data_path, 'train.csv')
test_file_path = os.path.join(processed_data_path, 'test.csv')

train_df = pd.read_csv(train_file_path)
test_df = pd.read_csv(test_file_path)
```

```python
# Plot histogram
plt.hist(train_df.Age)
```

```python
# Plot histogram and add some attributes
plt.hist(train_df.Age, bins=20, color='c')
plt.title('Age Histogram')
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.show()
```

```python
# Alternatively
f , ax = plt.subplots()
ax.hist(train_df.Age, bins=20, color='c')
ax.set_title('Age Histogram')
ax.set_xlabel('Bins')
ax.set_ylabel('Counts')
plt.show()
```

```python
# Add two subplots
f , (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# first subplot
ax1.hist(train_df.Age, bins=20, color='c')
ax1.set_title('Age Histogram')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Counts')

# second subplot
ax2.hist(train_df.Fare, bins=20, color='tomato')
ax2.set_title('Fare Histogram')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Counts')

plt.show()
```

```python
# Add more subplots
f , ax_arr = plt.subplots(3, 2, figsize=(10, 8))

# first subplot
ax_arr[0,0].hist(train_df.Age, bins=20, color='c')
ax_arr[0,0].set_title('Age Histogram')
ax_arr[0,0].set_xlabel('Bins')
ax_arr[0,0].set_ylabel('Counts')

# second subplot
ax_arr[0,1].hist(train_df.Fare, bins=20, color='tomato')
ax_arr[0,1].set_title('Fare Histogram')
ax_arr[0,1].set_xlabel('Bins')
ax_arr[0,1].set_ylabel('Counts')

# Thrid subplot
ax_arr[1,0].boxplot(train_df.Age.values)
ax_arr[1,0].set_title('Age Boxplot')
ax_arr[1,0].set_xlabel('Age')
ax_arr[1,0].set_ylabel('Age')

# Forth subplot
ax_arr[1,1].boxplot(train_df.Fare.values)
ax_arr[1,1].set_title('Fare Boxplot')
ax_arr[1,1].set_xlabel('Fare')
ax_arr[1,1].set_ylabel('Fare')

# Fifth subplot
ax_arr[2,0].scatter(train_df.Age, train_df.Fare, color='c', alpha=0.2)
ax_arr[2,0].set_title('Age vs Fare scatter plot')
ax_arr[2,0].set_xlabel('Age')
ax_arr[2,0].set_ylabel('Fare')

# remove empty frame
ax_arr[2,1].axis('off')
# fix overlap
plt.tight_layout()

plt.show()
```
